```python
import pandas as pd
import matplotlib.pyplot as plt

# Load & aggregate (Date as index)
supplement_sales = pd.read_csv(
    "/Users/vernesapodrimaj/Downloads/Supplement_Sales_Weekly_Expanded.csv",
    parse_dates=['Date'],
    index_col='Date'
)
weekly = supplement_sales.groupby(level=0)[['Revenue','Units Sold']].sum()

# Plot with secondary y-axis
ax = weekly.plot(
    secondary_y='Units Sold',
    figsize=(10,5),
    title='Weekly Revenue and Units Sold Over Time'
)
ax.set_xlabel('Date')
ax.set_ylabel('Revenue (€)')
ax.right_ax.set_ylabel('Units Sold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Use your real dataset
# supplement_sales = pd.read_csv('your_file.csv')  # Assuming it's already loaded

# Aggregate Units Sold and Revenue by Product Name
totals = supplement_sales.groupby('Product Name').agg({'Units Sold': 'sum', 'Revenue': 'sum'}).reset_index()

# Round Revenue to 2 decimals
totals['Revenue'] = totals['Revenue'].round(2)

# Sort by Revenue descending
totals_sorted = totals.sort_values('Revenue', ascending=False)

# Take the Top 10 Products
top10 = totals_sorted.head(10)

# Show Top 10 Products
print(top10)

# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x='Revenue', y='Product Name', palette='viridis')

# Labels and Title
plt.title('Top 10 Products by Total Revenue')
plt.xlabel('Revenue ($)')
plt.ylabel('Product Name')

# Layout
plt.tight_layout()

# Save figure
plt.savefig('top10_products_revenue.png')

# Show plot
plt.show()


```

```python
import pandas as pd

# Aggregate metrics by Category
category_summary = supplement_sales.groupby('Category').agg(
    Total_Revenue=('Revenue', 'sum'),
    Average_Price=('Price', 'mean'),
    Total_Units_Sold=('Units Sold', 'sum')
).reset_index()

# Round Revenue and Price
category_summary['Total_Revenue'] = category_summary['Total_Revenue'].round(2)
category_summary['Average_Price'] = category_summary['Average_Price'].round(2)

# Sort by Total Revenue descending
category_summary = category_summary.sort_values('Total_Revenue', ascending=False)

# Show the result
print(category_summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Re-compute and display platform summary
totals_platform = supplement_sales.groupby('Platform').agg(
    Revenue=('Revenue', 'sum'),
    Units_Sold=('Units Sold', 'sum')
).reset_index()

totals_platform['Revenue'] = totals_platform['Revenue'].round(2)
totals_platform = totals_platform.sort_values('Revenue', ascending=False)

print(totals_platform)

# Plot
plt.figure(figsize=(12, 5))

# Revenue plot
plt.subplot(1, 2, 1)
sns.barplot(data=totals_platform, x='Platform', y='Revenue', palette='rocket')
plt.title('Total Revenue by Platform')
plt.xlabel('Platform')
plt.ylabel('Revenue')

# Units Sold plot
plt.subplot(1, 2, 2)
sns.barplot(data=totals_platform, x='Platform', y='Units_Sold', palette='mako')
plt.title('Total Units Sold by Platform')
plt.xlabel('Platform')
plt.ylabel('Units Sold')

plt.tight_layout()
plt.savefig('platform_analysis_fixed.png')
plt.show()
```
